```python
FILE_PATH = "../data/Wave.opt"
with open(FILE_PATH, "r") as f:
    text = f.read()
blocks = re.split(r'(?=TIME\s*=)', text)

blocks = [b.strip() for b in blocks if b.strip()]
```

```python
def robust_peak(samples, n=10):
    top = sorted(samples)[-n:]
    return np.median(top)
```

```python
records = []

for block in blocks:

    lines = block.splitlines()

    header = lines[0]

    # Extract metadata
    m = re.search(
        r'TIME\s*=\s*([^,]+),\s*NR\s*=\s*(\d+),\s*SEQ\s*=\s*(\d+),\s*TS\[us\]\s*=\s*(\d+),\s*LOST_TRIG\s*=\s*(\d+)',
        header
    )

    time = m.group(1)
    nr = int(m.group(2))
    seq = int(m.group(3))
    ts_us = int(m.group(4))
    lost_trig = int(m.group(5))

    # Remaining lines are sample values
    samples_text = " ".join(lines[1:])

    samples = [
        float(x.strip())
        for x in samples_text.split(",")
        if x.strip()
    ]

    records.append({
        "time": time,
        "nr": nr,
        "seq": seq,
        "ts_us": ts_us,
        "lost_trig": lost_trig,
        "MaxSamples": robust_peak(samples)
    })

    # df["peak"] = df["samples"].apply(robust_peak)
```

```python
df = pd.DataFrame(records)
# df["time"] = pd.to_datetime(df["time"], format="%H:%M:%S")
# df["timestamp"] = (
#     df["time"]
#     + pd.to_timedelta(df["ts_us"], unit="us")
# )

base_date = "2026-06-07"  # or any reference date
df["datetime"] = pd.to_datetime(base_date + " " + df["time"])

df["datetime"] = df["datetime"] + pd.to_timedelta(df["ts_us"], unit="us")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_peak_between_times(df, start_time, end_time, base_date="2026-06-07"):
    """
    Filters dataframe between two HH:MM:SS timestamps and plots peak vs time.
    """

    # Build full datetime (including ts_us)
    if "datetime" not in df.columns:
        df = df.copy()
        df["datetime"] = pd.to_datetime(base_date + " " + df["time"])
        df["datetime"] += pd.to_timedelta(df["ts_us"], unit="us")

    # Convert window
    start = pd.to_datetime(base_date + " " + start_time)
    end = pd.to_datetime(base_date + " " + end_time)

    # Filter
    df_window = df[(df["datetime"] >= start) & (df["datetime"] <= end)]

    # Plot
    fig, ax = plt.subplots(figsize=(12, 5))
    ax.plot(df_window["datetime"], df_window["MaxSamples"])

    ax.set_xlabel("Time")
    ax.set_ylabel("Amplitude")
    ax.set_title(f"Peak between {start_time} and {end_time}")

    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    return df_window
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_peak_time_series(records, start_time=None, end_time=None, window=10):
    """
    records: list of dicts with keys:
        time (str), nr, ts_us, samples, peak
    start_time/end_time: strings like "11:39:35"
    window: smoothing window for rolling mean
    """

    df = pd.DataFrame(records)

    # ---- Convert time ----
    # combine TIME + ts_us into a sortable timestamp
    df["time_full"] = pd.to_datetime(
        df["time"], format="%H:%M:%S"
    ) + pd.to_timedelta(df["ts_us"], unit="us")

    # ---- Set NR as index ----
    df = df.sort_values("nr").set_index("nr")

    # ---- Filter by time range (optional) ----
    if start_time:
        start = pd.to_datetime(start_time, format="%H:%M:%S")
        df = df[df["time_full"] >= start]

    if end_time:
        end = pd.to_datetime(end_time, format="%H:%M:%S")
        df = df[df["time_full"] <= end]

    # ---- Smooth peak ----
    df["peak_smooth"] = df["MaxSamples"].rolling(window=window, center=True).mean()

    # ---- Plot ----
    fig, ax = plt.subplots(figsize=(12, 5))

    ax.plot(df["time_full"], df["MaxSamples"], alpha=0.3, label="Raw Peak")
    ax.plot(df["time_full"], df["peak_smooth"], linewidth=2, label="Smoothed Peak")

    # remove date from x-axis display
    ax.xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter("%H:%M:%S"))

    ax.set_xlabel("Time (HH:MM:SS)")
    ax.set_ylabel("Amplitude")
    ax.legend()
    ax.grid(True)

    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    return df
```

```python
import matplotlib.pyplot as plt

def plot_raw_and_smooth(df):
    """
    Plot raw and smoothed peaks in two separate subplots.

    Parameters
    ----------
    df : pandas.DataFrame
        DataFrame containing:
        - time_full
        - MaxSamples
        - peak_smooth
    """

    fig, (ax1, ax2) = plt.subplots(
        2, 1, figsize=(12, 8), sharex=True
    )

    # Raw Peak
    ax1.plot(df["time_full"], df["MaxSamples"], alpha=0.7)
    ax1.set_title("Raw Peak")
    ax1.set_ylabel("Amplitude")
    ax1.grid(True)

    # Smoothed Peak
    ax2.plot(df["time_full"], df["peak_smooth"], linewidth=2)
    ax2.set_title("Smoothed Peak")
    ax2.set_xlabel("Time (HH:MM:SS)")
    ax2.set_ylabel("Amplitude")
    ax2.grid(True)

    # Format x-axis
    ax2.xaxis.set_major_formatter(
        plt.matplotlib.dates.DateFormatter("%H:%M:%S")
    )

    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_selected_intervals(df, intervals):
    """
    Plots multiple selected time intervals together in a single continuous plot.

    intervals: List of tuples/lists containing (start_time, end_time)
    e.g., [("11:39:45", "11:41:30"), ("11:44:10", "11:45:50")]
    """
    # Ensure time_full is in datetime format for accurate filtering
    # Adjust format string if your dataframe includes dates
    df_working = df.copy()
    df_working["time_datetime"] = pd.to_datetime(df_working["time_full"], format="%H:%M:%S")

    extracted_chunks = []

    # Loop through each selected interval and grab the matching data
    for start_str, end_str in intervals:
        start_dt = pd.to_datetime(start_str, format="%H:%M:%S")
        end_dt = pd.to_datetime(end_str, format="%H:%M:%S")

        chunk = df_working[(df_working["time_datetime"] >= start_dt) &
        (df_working["time_datetime"] <= end_dt)]
        extracted_chunks.append(chunk)

    if not extracted_chunks:
        print("No intervals selected.")
        return

        # Combine all selected intervals into one continuous DataFrame
    combined_df = pd.concat(extracted_chunks).drop_duplicates().sort_values("time_datetime")

        # --- Plotting both on a single layout ---
    plt.figure(figsize=(12, 6))

        # 1. Plot Raw Data (thinner/transparent)
    plt.plot(combined_df["time_full"], combined_df["MaxSamples"],
    alpha=0.4, color="gray", label="Raw Peak")

        # 2. Plot Smoothed Data (bolder)
    plt.plot(combined_df["time_full"], combined_df["peak_smooth"],
    linewidth=2, color="blue", label="Smoothed Peak")

        # Formatting
    plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter("%H:%M:%S"))
    plt.title("PP (MFI 11) 200°C - Selected Intervals")
    plt.xlabel("Time (HH:MM:SS)")
    plt.ylabel("Amplitude")

    plt.legend(loc="upper right")
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.xticks(rotation=45)
    plt.tight_layout()

    plt.show()

# ==========================================
# HOW TO CALL THE FUNCTION WITH YOUR NOTES:
# ==========================================
# You can cherry-pick whichever numbers (1 to 10) you want from your notebook here:
my_selected_intervals = [
("11:39:45", "11:41:30"), # Interval 1
("11:42:00", "11:43:40"), # Interval 2
("11:44:10", "11:45:50"), # Interval 3
("11:46:20", "11:48:00"), # Interval 4
("11:48:30", "11:50:10"), # Interval 5
("11:50:40", "11:52:20"), # Interval 6
("11:52:50", "11:54:25"), # Interval 7
("11:55:00", "11:56:40"), # Interval 8
("11:57:10", "11:58:40"), # Interval 9
("11:59:20", "12:00:55"), # Interval 10
# Add or remove any of the 10 intervals here as needed...
]

plot_selected_intervals(df, my_selected_intervals)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_selected_intervals_separated(df, intervals):
    """
    Plots multiple selected time intervals into two completely separate graphs:
    one for Raw Peak and one for Smoothed Peak.
    """
    df_working = df.copy()
    df_working["time_datetime"] = pd.to_datetime(df_working["time_full"], format="%H:%M:%S")
    
    extracted_chunks = []
    
    # Loop through each selected interval and grab the matching data
    for start_str, end_str in intervals:
        start_dt = pd.to_datetime(start_str, format="%H:%M:%S")
        end_dt = pd.to_datetime(end_str, format="%H:%M:%S")
        
        chunk = df_working[(df_working["time_datetime"] >= start_dt) & 
                           (df_working["time_datetime"] <= end_dt)]
        extracted_chunks.append(chunk)
    
    if not extracted_chunks:
        print("No intervals selected.")
        return
        
    # Combine all selected intervals into one continuous DataFrame
    combined_df = pd.concat(extracted_chunks).drop_duplicates().sort_values("time_datetime")

    # =========================================================================
    # GRAPH 1: RAW PEAK ONLY
    # =========================================================================
    plt.figure(figsize=(12, 5))
    plt.plot(combined_df["time_full"], combined_df["MaxSamples"], 
             alpha=0.7, color="orange")  # Increased opacity since it's on its own now
    
    plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter("%H:%M:%S"))
    plt.title("PP (MFI 11) 200°C Raw Peak - Selected Intervals")
    plt.xlabel("Time (HH:MM:SS)")
    plt.ylabel("Amplitude")
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.xticks(rotation=45)
    plt.tight_layout()
    
    # =========================================================================
    # GRAPH 2: SMOOTHED PEAK ONLY (Placing this in a completely separate window)
    # =========================================================================
    plt.figure(figsize=(12, 5))
    plt.plot(combined_df["time_full"], combined_df["peak_smooth"], 
             linewidth=2, color="blue")
    
    plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter("%H:%M:%S"))
    plt.title("PP (MFI 11) 200°C Smoothed Peak - Selected Intervals")
    plt.xlabel("Time (HH:MM:SS)")
    plt.ylabel("Amplitude")
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.xticks(rotation=45)
    plt.tight_layout()
    
    # Display both separate windows at the same time
    plt.show()
```
