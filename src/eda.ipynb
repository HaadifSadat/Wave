```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
opBox200 = pd.read_csv("..\data\Wave.opt")
opBox200.head(10)
```

```python
FILE_PATH = "..\data\Wave.opt"
with open(FILE_PATH, "r") as f:
    text = f.read()
blocks = re.split(r'(?=TIME\s*=)', text)

blocks = [b.strip() for b in blocks if b.strip()]
```

```python
records = []

for block in blocks:

    lines = block.splitlines()

    header = lines[0]

    # Extract metadata
    m = re.search(
        r'TIME\s*=\s*([^,]+),\s*NR\s*=\s*(\d+),\s*SEQ\s*=\s*(\d+),\s*TS\[us\]\s*=\s*(\d+),\s*LOST_TRIG\s*=\s*(\d+)',
        header
    )

    time = m.group(1)
    nr = int(m.group(2))
    seq = int(m.group(3))
    ts_us = int(m.group(4))
    lost_trig = int(m.group(5))

    # Remaining lines are sample values
    samples_text = " ".join(lines[1:])

    samples = [
        float(x.strip())
        for x in samples_text.split(",")
        if x.strip()
    ]

    records.append({
        "time": time,
        "nr": nr,
        "seq": seq,
        "ts_us": ts_us,
        "lost_trig": lost_trig,
        "samples": samples
    })
```

```python
df = pd.DataFrame(records)
```

```python
rows = []

for rec in records:
    for idx, value in enumerate(rec["samples"]):

        rows.append({
            "nr": rec["nr"],
            "sample_idx": idx,
            "value": value,
            "ts_us": rec["ts_us"]
        })

samples_df = pd.DataFrame(rows)

samples_df.head()
```
